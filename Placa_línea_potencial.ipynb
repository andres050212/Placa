```python
from numpy import empty, zeros, max
from pylab import imshow,gray,show,inferno,cool,plasma,magma
```

```python
M=100 
v=1.0  
target= 1e-12



phi= zeros([M+1,M+1],float)
phi[0,:]=0
rho = zeros([M+1,M+1],float)
rho[20:80,80:81] = v
rho[20:80,20:21] = -v  
```

```python
delta = 1.0
omega = 0.9
while delta>target:
    delta = 0
    for i in range(1,M):
        for j in range(1,M):
            difference=(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])/4 - phi[i,j]+rho[i,j]
            phi[i,j] = phi[i,j]+(1+omega)*difference
        
        if difference>delta: delta = difference

imshow(phi)
inferno()
show()
```
