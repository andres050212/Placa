```python
from numpy import empty,zeros,max
from pylab import imshow,gray,inferno,show
```

```python
###### Constantes ######
M = 100
V = 1.0 
target = 1e-10
########################
```

```python
###### Vector para guardar los valores del potencial ######
phi=zeros([M+1,M+1],float)
phi[0,:]=V
###########################################################

###### Ciclo para calcular los valores del potencial ######
delta = 1.0
omega = 0.9
while delta>target:
    delta = 0
    for i in range(1,M):
        for j in range(1,M):
            difference=(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])/4 - phi[i,j]
            phi[i,j] = phi[i,j]+(1+omega)*difference
        
        if difference>delta: delta = difference
            
############################################################


###### Gr√°fica ######
imshow(phi)
inferno()
show()
```
